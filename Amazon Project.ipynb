```python
#import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
```

```python
#load Dataset
df = pd.read_csv('amazon.csv')
df.head()
```

```python
#check for missing values
df.isnull().sum()
```

```python
# drop missing values as they are few
df = df.dropna()
```

```python
#check for missing values again to confirm they are dropped
df.isnull().sum()
```

```python
#check for duplicates
print(df.duplicated().sum())
```

```python
#display Columns names
df.columns
```

```python
# Show datatypes of all columns
print(df.dtypes)
```

```python
#drop unnecessary columns
df = df.drop(columns=['product_id', 'about_product', 'user_id', 'user_name', 'review_id', 'img_link', 'product_link','review_content','review_title'])
df.head()
```

```python
# Remove ₹, commas, and % then convert to numeric
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '').astype(float)
df['rating_count'] = (df['rating_count'].astype(str).str.replace(',', '', regex=False).astype(int))
df.head()
```

```python
df['category'].value_counts()
```

```python
# Extract main category
df['main_category'] = df['category'].str.split('|').str[0]
df['main_category'].value_counts()
```

```python
# Check the unique values in the 'rating' column as '|' noticed while trying to convert it to numeric
df['rating'].value_counts()
```

```python
# Convert 'rating' to numeric, setting errors='coerce' to handle non-numeric values and convert them to NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
#drop missing values again after converting 'rating' to numeric
df = df.dropna()
```

```python
# Exploratory Data Analysis (EDA)
df.describe()
```

```python
df.info()
```

Q1- Which categories have the highest average ratings?

```python
total_ratings_by_category = df.groupby('main_category')['rating'].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=total_ratings_by_category.values, y=total_ratings_by_category.index, palette='viridis')
plt.title("Top 10 Categories by Rating")
plt.xlabel("Total Rating")
plt.ylabel("Category")
plt.show()
```

Electronics Category is the highest rating category

Q2- What are the top 5 products have the highest rating counts (most popular)?

```python
# Top 5 products with highest rating counts
top_products = df[['product_name','rating_count']].sort_values(by="rating_count", ascending=False).head(5)
top_products
```

```python
# Visualize top 5 products by rating count
plt.figure(figsize=(12,6))
plt.pie(top_products['rating_count'],labels=top_products['product_name'],autopct='%.2f%%')
plt.title("Top 10 Most Reviewed Products")
plt.xlabel("Number of Reviews")
plt.ylabel("Product")
plt.show()
```

Q3- What’s the average discount percentage across categories?

```python
avg_discount = df.groupby('main_category')['discount_percentage'].mean().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=avg_discount.values, y=avg_discount.index, palette='viridis')
plt.title("Top 10 Categories by Average Discount %")
plt.xlabel("Average Discount %")
plt.ylabel("Category")
plt.show()
```

Home Improvement Category has the highest average Discount %

Q4- Are higher priced products rated better or worse? and  What is the highest factor that affects rating & review?

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="actual_price", y="rating")
plt.title("Relationship Between Product Price and Rating")
plt.xlabel("Price (Actual)")
plt.ylabel("Rating")
plt.show()
```

```python
print("Correlation between actual price and rating:", df['actual_price'].corr(df['rating']))
```

Ratings don’t strongly increase with price

There is no clear correlation between higher price and better rating

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="discount_percentage", y="rating")
plt.title("Do Bigger Discounts Lead to Higher Ratings?")
plt.xlabel("Discount Percentage (%)")
plt.ylabel("Rating")
plt.show()
```

```python
print("Correlation between discount % and rating:", df['discount_percentage'].corr(df['rating']))
```

Discounts doesn’t affect rating much.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="discounted_price", y="rating")
plt.title("Do  Discount price Lead to Higher Ratings?")
plt.xlabel("Discount price")
plt.ylabel("Rating")
plt.show()
```

```python
print("Correlation between discounted_price and rating:", df['discounted_price'].corr(df['rating']))
```

Dicounted price doesn’t affect rating much so does discount percentage and actual price
